```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Mock dataset for demonstration (replace with a larger, real dataset)
data = {
    'food_name': [
        'Spaghetti Bolognese', 'Chicken Curry', 'Veggie Salad', 'Pancakes', 'Grilled Sandwich',
        'Fruit Smoothie', 'Burger', 'Pizza', 'Sushi', 'Pasta', 'Tacos', 'Burrito', 'Ramen',
        'Salmon', 'Chicken Salad', 'Steak', 'Grilled Cheese', 'Fruit Salad', 'Lasagna', 'Omelette',
        'Chicken Alfredo', 'Vegetable Stir Fry', 'Fish Tacos', 'Beef Stroganoff', 'Avocado Toast',
        'Clam Chowder', 'Eggplant Parmesan', 'Shrimp Scampi', 'Caesar Salad', 'French Onion Soup',
        'Stuffed Bell Peppers', 'Chicken Noodle Soup', 'Beef Wellington', 'Caprese Salad',
        'BBQ Ribs', 'Sweet Potato Fries', 'Lentil Soup', 'Mushroom Risotto', 'Cobb Salad', 'Falafel Wrap',
        'Eggplant Parmesan', 'Seafood Paella', 'Kimchi Fried Rice', 'Shakshuka', 'Beetroot Salad',
        'Lamb Kebab', 'Cajun Jambalaya', 'Pumpkin Soup', 'Paneer Butter Masala', 'Garlic Shrimp',
        'Stuffed Zucchini', 'Poke Bowl', 'Clam Chowder', 'Duck Confit', 'Lobster Roll',
        'Bibimbap', 'Tempura', 'Minestrone Soup', 'Croque Madame', 'Gnocchi'
    ],
    'ingredients': [
        "spaghetti, ground beef, tomato sauce", "chicken breast, coconut milk, spices", "lettuce, olive oil, tomatoes",
        "flour, eggs, milk, sugar", "bread, butter, cheese", "banana, milk, honey",
        "beef patty, lettuce, tomato, cheese, bun", "flour, yeast, cheese, tomato sauce",
        "fish, rice, seaweed, vegetables", "pasta, tomato sauce, garlic, herbs", "taco shell, beef, lettuce, cheese, salsa",
        "tortilla, beans, cheese, rice, beef", "noodles, broth, soy sauce, vegetables", "salmon, herbs, lemon",
        "chicken, lettuce, tomatoes, olive oil", "steak, potatoes, butter, herbs", "bread, cheese, butter, herbs",
        "fruits, honey, yogurt", "pasta, cheese, tomato sauce, herbs", "eggs, milk, butter, cheese",
        "chicken breast, pasta, cream, parmesan", "mixed vegetables, soy sauce, garlic, ginger",
        "fish fillets, tortillas, coleslaw, salsa", "beef, mushrooms, sour cream, egg noodles",
        "bread, avocado, olive oil, spices", "clams, potatoes, cream, celery, onion",
        "eggplant, marinara sauce, mozzarella, breadcrumbs", "shrimp, butter, garlic, pasta",
        "romaine lettuce, croutons, Caesar dressing", "onions, beef broth, croutons, gruyere cheese",
        "bell peppers, rice, ground beef, tomato sauce", "chicken, noodles, carrots, celery, broth",
        "beef tenderloin, puff pastry, mushrooms, dijon mustard", "tomatoes, mozzarella, basil, olive oil",
        "pork ribs, BBQ sauce, spices", "sweet potatoes, oil, spices",
        "lentils, carrots, celery, onions, broth", "arborio rice, mushrooms, parmesan, white wine",
        "lettuce, chicken, eggs, avocado, blue cheese", "chickpeas, flour, spices, pita bread, tahini",
        "eggplant, tomato sauce, mozzarella, breadcrumbs", "rice, saffron, shrimp, mussels, chicken",
        "kimchi, rice, soy sauce, sesame oil", "tomatoes, eggs, bell peppers, spices",
        "beetroot, arugula, goat cheese, walnuts", "lamb, onions, spices, yogurt",
        "rice, sausage, shrimp, Cajun spices", "pumpkin, cream, spices, broth",
        "paneer, butter, tomato puree, cream", "shrimp, garlic, butter, parsley",
        "zucchini, quinoa, cheese, breadcrumbs", "sushi rice, tuna, avocado, soy sauce",
        "clams, potatoes, cream, bacon", "duck, garlic, thyme, duck fat",
        "lobster, butter, mayonnaise, bread roll", "rice, beef, vegetables, gochujang",
        "shrimp, tempura batter, vegetables, soy sauce", "pasta, beans, tomatoes, vegetables",
        "bread, ham, cheese, egg, mustard", "potatoes, flour, egg, butter"

    ],
    'nutrition': [
        {"calories": 500, "carbohydrates": 70, "proteins": 25, "fats": 15, "vitamins": 10, "minerals": 5},
        {"calories": 400, "carbohydrates": 30, "proteins": 40, "fats": 10, "vitamins": 20, "minerals": 10},
        {"calories": 150, "carbohydrates": 10, "proteins": 5, "fats": 10, "vitamins": 30, "minerals": 10},
        {"calories": 300, "carbohydrates": 50, "proteins": 10, "fats": 10, "vitamins": 5, "minerals": 5},
        {"calories": 400, "carbohydrates": 50, "proteins": 15, "fats": 25, "vitamins": 5, "minerals": 5},
        {"calories": 250, "carbohydrates": 60, "proteins": 5, "fats": 5, "vitamins": 10, "minerals": 10},
        {"calories": 600, "carbohydrates": 50, "proteins": 30, "fats": 25, "vitamins": 15, "minerals": 10},
        {"calories": 700, "carbohydrates": 60, "proteins": 30, "fats": 30, "vitamins": 20, "minerals": 15},
        {"calories": 250, "carbohydrates": 35, "proteins": 15, "fats": 8, "vitamins": 10, "minerals": 5},
        {"calories": 450, "carbohydrates": 55, "proteins": 25, "fats": 15, "vitamins": 12, "minerals": 8},
        {"calories": 350, "carbohydrates": 40, "proteins": 15, "fats": 10, "vitamins": 10, "minerals": 8},
        {"calories": 550, "carbohydrates": 60, "proteins": 25, "fats": 20, "vitamins": 10, "minerals": 5},
        {"calories": 700, "carbohydrates": 75, "proteins": 30, "fats": 25, "vitamins": 15, "minerals": 10},
        {"calories": 300, "carbohydrates": 50, "proteins": 10, "fats": 10, "vitamins": 5, "minerals": 5},
        {"calories": 450, "carbohydrates": 50, "proteins": 30, "fats": 20, "vitamins": 15, "minerals": 10},
        {"calories": 550, "carbohydrates": 60, "proteins": 40, "fats": 30, "vitamins": 20, "minerals": 15},
        {"calories": 350, "carbohydrates": 40, "proteins": 15, "fats": 15, "vitamins": 12, "minerals": 8},
        {"calories": 300, "carbohydrates": 45, "proteins": 20, "fats": 15, "vitamins": 10, "minerals": 7},
        {"calories": 400, "carbohydrates": 50, "proteins": 25, "fats": 20, "vitamins": 15, "minerals": 8},
        {"calories": 500, "carbohydrates": 60, "proteins": 30, "fats": 20, "vitamins": 15, "minerals": 10},
        {"calories": 600, "carbohydrates": 50, "proteins": 40, "fats": 25, "vitamins": 10, "minerals": 8},
        {"calories": 200, "carbohydrates": 30, "proteins": 10, "fats": 5, "vitamins": 25, "minerals": 10},
        {"calories": 300, "carbohydrates": 35, "proteins": 25, "fats": 10, "vitamins": 15, "minerals": 8},
        {"calories": 550, "carbohydrates": 45, "proteins": 35, "fats": 20, "vitamins": 8, "minerals": 10},
        {"calories": 350, "carbohydrates": 40, "proteins": 10, "fats": 20, "vitamins": 12, "minerals": 5},
        {"calories": 400, "carbohydrates": 35, "proteins": 25, "fats": 20, "vitamins": 8, "minerals": 12},
        {"calories": 450, "carbohydrates": 50, "proteins": 15, "fats": 20, "vitamins": 10, "minerals": 8},
        {"calories": 500, "carbohydrates": 60, "proteins": 35, "fats": 25, "vitamins": 12, "minerals": 10},
        {"calories": 300, "carbohydrates": 25, "proteins": 15, "fats": 10, "vitamins": 10, "minerals": 5},
        {"calories": 350, "carbohydrates": 30, "proteins": 10, "fats": 15, "vitamins": 5, "minerals": 8},
        {"calories": 400, "carbohydrates": 50, "proteins": 25, "fats": 15, "vitamins": 10, "minerals": 8},
        {"calories": 200, "carbohydrates": 30, "proteins": 20, "fats": 5, "vitamins": 10, "minerals": 8},
        {"calories": 700, "carbohydrates": 40, "proteins": 45, "fats": 35, "vitamins": 8, "minerals": 12},
        {"calories": 250, "carbohydrates": 10, "proteins": 10, "fats": 15, "vitamins": 15, "minerals": 5},
        {"calories": 600, "carbohydrates": 30, "proteins": 35, "fats": 25, "vitamins": 5, "minerals": 10},
        {"calories": 300, "carbohydrates": 45, "proteins": 5, "fats": 10, "vitamins": 8, "minerals": 5},
        {"calories": 250, "carbohydrates": 35, "proteins": 10, "fats": 5, "vitamins": 20, "minerals": 12},
        {"calories": 500, "carbohydrates": 40, "proteins": 20, "fats": 15, "vitamins": 10, "minerals": 12},
        {"calories": 600, "carbohydrates": 30, "proteins": 35, "fats": 20, "vitamins": 10, "minerals": 15},
        {"calories": 350, "carbohydrates": 45, "proteins": 15, "fats": 10, "vitamins": 10, "minerals": 8},
        {"calories": 350, "carbohydrates": 40, "proteins": 15, "fats": 10, "vitamins": 12, "minerals": 8},
        {"calories": 550, "carbohydrates": 60, "proteins": 30, "fats": 15, "vitamins": 10, "minerals": 12},
        {"calories": 400, "carbohydrates": 55, "proteins": 12, "fats": 10, "vitamins": 8, "minerals": 6},
        {"calories": 250, "carbohydrates": 20, "proteins": 15, "fats": 10, "vitamins": 20, "minerals": 8},
        {"calories": 200, "carbohydrates": 15, "proteins": 5, "fats": 10, "vitamins": 25, "minerals": 10},
        {"calories": 450, "carbohydrates": 30, "proteins": 35, "fats": 15, "vitamins": 10, "minerals": 15},
        {"calories": 550, "carbohydrates": 55, "proteins": 25, "fats": 18, "vitamins": 12, "minerals": 8},
        {"calories": 300, "carbohydrates": 25, "proteins": 10, "fats": 15, "vitamins": 18, "minerals": 12},
        {"calories": 500, "carbohydrates": 40, "proteins": 20, "fats": 25, "vitamins": 15, "minerals": 10},
        {"calories": 350, "carbohydrates": 10, "proteins": 25, "fats": 15, "vitamins": 10, "minerals": 8},
        {"calories": 300, "carbohydrates": 30, "proteins": 12, "fats": 10, "vitamins": 15, "minerals": 5},
        {"calories": 400, "carbohydrates": 45, "proteins": 25, "fats": 10, "vitamins": 12, "minerals": 8},
        {"calories": 450, "carbohydrates": 40, "proteins": 20, "fats": 20, "vitamins": 10, "minerals": 10},
        {"calories": 500, "carbohydrates": 20, "proteins": 35, "fats": 30, "vitamins": 10, "minerals": 8},
        {"calories": 450, "carbohydrates": 35, "proteins": 20, "fats": 15, "vitamins": 15, "minerals": 8},
        {"calories": 550, "carbohydrates": 50, "proteins": 20, "fats": 25, "vitamins": 12, "minerals": 10},
        {"calories": 250, "carbohydrates": 20, "proteins": 10, "fats": 15, "vitamins": 15, "minerals": 10},
        {"calories": 350, "carbohydrates": 40, "proteins": 10, "fats": 15, "vitamins": 18, "minerals": 12},
        {"calories": 400, "carbohydrates": 50, "proteins": 25, "fats": 20, "vitamins": 20, "minerals": 12},
        {"calories": 300, "carbohydrates": 35, "proteins": 10, "fats": 15, "vitamins": 15, "minerals": 10}
    ],
    'substitutes': [
        "whole wheat spaghetti, ground turkey, tomato paste", "chicken thighs, skim milk, spices",
        "spinach, sunflower oil, tomatoes", "whole wheat flour, egg whites, almond milk, stevia",
        "multigrain bread, low-fat spread, vegan cheese", "avocado, almond milk, maple syrup",
        "turkey patty, lettuce, tomato, vegan cheese, bun", "gluten-free flour, vegan cheese, vegetables",
        "tofu, rice, seaweed, vegetables", "gluten-free pasta, vegetable broth, garlic, herbs",
        "corn tortillas, ground turkey, lettuce, salsa", "brown rice, beans, tofu, vegetables",
        "rice noodles, miso broth, tofu, vegetables", "tilapia, herbs, lemon", "turkey, lettuce, tomatoes, olive oil",
        "vegetarian steak, potatoes, mushrooms, herbs", "low-fat bread, cheese, butter, herbs",
        "almond milk, honey, soy yogurt", "zucchini, eggplant, mushrooms, cheese", "egg whites, almond milk, mushrooms",
        "tofu, zucchini noodles, low-fat cream", "tofu, tamari, sesame oil",
        "tofu, corn tortillas, guacamole", "mushrooms, almond milk, rice noodles",
        "gluten-free bread, hummus, cucumbers", "mushrooms, coconut milk, leek",
        "zucchini, marinara, vegan cheese", "tofu, olive oil, lemon zest, quinoa",
        "spinach, nuts, vinaigrette", "mushrooms, vegetable broth, vegan cheese",
        "quinoa, lentils, tomato sauce", "tempeh, soba noodles, bok choy",
        "portobello mushrooms, vegan puff pastry", "tofu, basil, cherry tomatoes",
        "jackfruit, BBQ sauce, coleslaw", "carrot fries, olive oil, spices",
        "red lentils, spinach, curry spices", "cauliflower rice, mushrooms, nutritional yeast",
        "tofu, kale, pumpkin seeds, vinaigrette", "sweet potato wrap, hummus, spinach",
        "zucchini slices, vegan cheese, almond meal", "brown rice, tofu, saffron",
        "cauliflower rice, kimchi, coconut aminos", "tofu scramble, bell peppers, tomato puree",
        "carrots, spinach, almond slivers, tahini", "veggie patty, yogurt dressing, mushrooms",
        "quinoa, shrimp, Cajun spices", "sweet potato, coconut milk, cinnamon",
        "tofu, cashew cream, tomato paste", "mushrooms, garlic, olive oil",
        "eggplant, brown rice, nutritional yeast", "couscous, avocado, mango, sesame oil",
        "corn chowder, potatoes, vegan cream", "chicken, olive oil, rosemary",
        "shrimp roll, avocado, lemon mayo", "quinoa, tofu, zucchini, sriracha",
        "zucchini tempura, sesame oil, soy sauce", "barley, beans, kale, herbs",
        "whole-grain bread, vegan cheese, spinach", "sweet potato gnocchi, almond butter"
    ],
    'substitute_nutrition': [
        {"calories": 450, "carbohydrates": 65, "proteins": 20, "fats": 10, "vitamins": 15, "minerals": 5},
        {"calories": 350, "carbohydrates": 25, "proteins": 35, "fats": 8, "vitamins": 25, "minerals": 15},
        {"calories": 120, "carbohydrates": 8, "proteins": 4, "fats": 8, "vitamins": 35, "minerals": 12},
        {"calories": 270, "carbohydrates": 40, "proteins": 8, "fats": 5, "vitamins": 10, "minerals": 8},
        {"calories": 350, "carbohydrates": 40, "proteins": 10, "fats": 15, "vitamins": 10, "minerals": 8},
        {"calories": 220, "carbohydrates": 55, "proteins": 4, "fats": 4, "vitamins": 15, "minerals": 10},
        {"calories": 500, "carbohydrates": 60, "proteins": 25, "fats": 20, "vitamins": 15, "minerals": 10},
        {"calories": 650, "carbohydrates": 65, "proteins": 30, "fats": 25, "vitamins": 18, "minerals": 12},
        {"calories": 200, "carbohydrates": 25, "proteins": 10, "fats": 5, "vitamins": 12, "minerals": 6},
        {"calories": 500, "carbohydrates": 60, "proteins": 25, "fats": 10, "vitamins": 15, "minerals": 8},
        {"calories": 600, "carbohydrates": 50, "proteins": 20, "fats": 25, "vitamins": 12, "minerals": 10},
        {"calories": 550, "carbohydrates": 55, "proteins": 25, "fats": 20, "vitamins": 18, "minerals": 10},
        {"calories": 650, "carbohydrates": 70, "proteins": 30, "fats": 25, "vitamins": 20, "minerals": 12},
        {"calories": 280, "carbohydrates": 40, "proteins": 15, "fats": 15, "vitamins": 10, "minerals": 7},
        {"calories": 350, "carbohydrates": 50, "proteins": 20, "fats": 18, "vitamins": 15, "minerals": 10},
        {"calories": 400, "carbohydrates": 50, "proteins": 20, "fats": 20, "vitamins": 18, "minerals": 10},
        {"calories": 450, "carbohydrates": 55, "proteins": 25, "fats": 22, "vitamins": 18, "minerals": 12},
        {"calories": 500, "carbohydrates": 60, "proteins": 30, "fats": 25, "vitamins": 20, "minerals": 15},
        {"calories": 600, "carbohydrates": 70, "proteins": 35, "fats": 30, "vitamins": 25, "minerals": 20},
        {"calories": 320, "carbohydrates": 45, "proteins": 15, "fats": 12, "vitamins": 8, "minerals": 7},
        {"calories": 400, "carbohydrates": 30, "proteins": 20, "fats": 15, "vitamins": 20, "minerals": 12},
        {"calories": 150, "carbohydrates": 20, "proteins": 8, "fats": 5, "vitamins": 30, "minerals": 10},
        {"calories": 200, "carbohydrates": 25, "proteins": 10, "fats": 10, "vitamins": 15, "minerals": 8},
        {"calories": 300, "carbohydrates": 35, "proteins": 20, "fats": 10, "vitamins": 10, "minerals": 8},
        {"calories": 180, "carbohydrates": 25, "proteins": 5, "fats": 10, "vitamins": 15, "minerals": 5},
        {"calories": 250, "carbohydrates": 30, "proteins": 10, "fats": 8, "vitamins": 12, "minerals": 10},
        {"calories": 300, "carbohydrates": 30, "proteins": 8, "fats": 10, "vitamins": 10, "minerals": 8},
        {"calories": 350, "carbohydrates": 40, "proteins": 25, "fats": 15, "vitamins": 12, "minerals": 10},
        {"calories": 200, "carbohydrates": 10, "proteins": 5, "fats": 8, "vitamins": 12, "minerals": 5},
        {"calories": 300, "carbohydrates": 25, "proteins": 5, "fats": 8, "vitamins": 10, "minerals": 8},
        {"calories": 250, "carbohydrates": 35, "proteins": 10, "fats": 8, "vitamins": 12, "minerals": 10},
        {"calories": 200, "carbohydrates": 20, "proteins": 15, "fats": 5, "vitamins": 12, "minerals": 8},
        {"calories": 400, "carbohydrates": 30, "proteins": 20, "fats": 15, "vitamins": 10, "minerals": 10},
        {"calories": 180, "carbohydrates": 15, "proteins": 10, "fats": 8, "vitamins": 15, "minerals": 5},
        {"calories": 400, "carbohydrates": 25, "proteins": 15, "fats": 12, "vitamins": 8, "minerals": 5},
        {"calories": 150, "carbohydrates": 20, "proteins": 5, "fats": 5, "vitamins": 15, "minerals": 5},
        {"calories": 200, "carbohydrates": 30, "proteins": 8, "fats": 5, "vitamins": 20, "minerals": 12},
        {"calories": 300, "carbohydrates": 20, "proteins": 10, "fats": 5, "vitamins": 12, "minerals": 8},
        {"calories": 400, "carbohydrates": 25, "proteins": 20, "fats": 15, "vitamins": 15, "minerals": 12},
        {"calories": 250, "carbohydrates": 30, "proteins": 10, "fats": 10, "vitamins": 10, "minerals": 8},
        {"calories": 300, "carbohydrates": 35, "proteins": 10, "fats": 12, "vitamins": 20, "minerals": 10},
        {"calories": 500, "carbohydrates": 45, "proteins": 15, "fats": 10, "vitamins": 15, "minerals": 8},
        {"calories": 350, "carbohydrates": 40, "proteins": 12, "fats": 10, "vitamins": 18, "minerals": 10},
        {"calories": 250, "carbohydrates": 20, "proteins": 12, "fats": 10, "vitamins": 15, "minerals": 8},
        {"calories": 200, "carbohydrates": 10, "proteins": 8, "fats": 15, "vitamins": 25, "minerals": 10},
        {"calories": 400, "carbohydrates": 30, "proteins": 20, "fats": 12, "vitamins": 18, "minerals": 12},
        {"calories": 450, "carbohydrates": 40, "proteins": 20, "fats": 15, "vitamins": 10, "minerals": 5},
        {"calories": 280, "carbohydrates": 30, "proteins": 10, "fats": 12, "vitamins": 20, "minerals": 8},
        {"calories": 450, "carbohydrates": 35, "proteins": 15, "fats": 25, "vitamins": 12, "minerals": 8},
        {"calories": 300, "carbohydrates": 15, "proteins": 20, "fats": 10, "vitamins": 10, "minerals": 5},
        {"calories": 250, "carbohydrates": 30, "proteins": 15, "fats": 8, "vitamins": 15, "minerals": 8},
        {"calories": 380, "carbohydrates": 40, "proteins": 20, "fats": 10, "vitamins": 18, "minerals": 10},
        {"calories": 400, "carbohydrates": 35, "proteins": 15, "fats": 12, "vitamins": 10, "minerals": 8},
        {"calories": 480, "carbohydrates": 20, "proteins": 25, "fats": 25, "vitamins": 12, "minerals": 8},
        {"calories": 420, "carbohydrates": 30, "proteins": 20, "fats": 15, "vitamins": 12, "minerals": 10},
        {"calories": 550, "carbohydrates": 45, "proteins": 20, "fats": 20, "vitamins": 15, "minerals": 10},
        {"calories": 250, "carbohydrates": 20, "proteins": 12, "fats": 15, "vitamins": 20, "minerals": 12},
        {"calories": 350, "carbohydrates": 30, "proteins": 8, "fats": 12, "vitamins": 18, "minerals": 10},
        {"calories": 400, "carbohydrates": 50, "proteins": 15, "fats": 15, "vitamins": 20, "minerals": 12},
        {"calories": 300, "carbohydrates": 30, "proteins": 10, "fats": 12, "vitamins": 12, "minerals": 10}
    ],

}

# Convert data into a DataFrame
df = pd.DataFrame(data)

# Function to fetch ingredients and nutrition
def fetch_nutrition(food_name):
    if food_name in df['food_name'].values:
        idx = df[df['food_name'] == food_name].index[0]
        return df.iloc[idx]['ingredients'], df.iloc[idx]['nutrition']
    return None, None

# Function to substitute ingredients
def substitute_ingredients(food_name):
    if food_name in df['food_name'].values:
        idx = df[df['food_name'] == food_name].index[0]
        return df.iloc[idx]['substitutes'], df.iloc[idx]['substitute_nutrition']
    return None, None

# Function to display nutrition comparison
def display_comparison(original, substitute=None):
    categories = ['calories', 'carbohydrates', 'proteins', 'fats', 'vitamins', 'minerals']

    # Create a DataFrame for displaying
    table_data = {
        'Category': categories,
        'Original': [original[cat] for cat in categories],
        'Substitute' if substitute else 'N/A': [substitute[cat] if substitute else 'N/A' for cat in categories]
    }
    df_table = pd.DataFrame(table_data)

    # Plot comparison if substitution is performed
    if substitute:
        original_values = [original[cat] for cat in categories]
        substitute_values = [substitute[cat] for cat in categories]

        # Create a bar graph
        fig, ax = plt.subplots(figsize=(12, 6))
        width = 0.35  # Bar width
        x = np.arange(len(categories))

        ax.bar(x - width/2, original_values, width, label='Original', color='#4CAF50')  # Dark Green
        ax.bar(x + width/2, substitute_values, width, label='Substitute', color='#2196F3')  # Blue

        ax.set_xlabel('Nutritional Categories')
        ax.set_ylabel('Values')
        ax.set_title('Nutritional Comparison')
        ax.set_xticks(x)
        ax.set_xticklabels(categories, rotation=45, ha='right')
        ax.legend()

        plt.tight_layout()
        plt.show()

        # Create a pie chart for the original food
        fig, ax = plt.subplots(figsize=(8, 8))  # Increased size for pie chart

        ax.pie(
            original_values,
            startangle=90,
            wedgeprops={'edgecolor': 'black', 'linewidth': 1},
            colors=['#f1c40f', '#e67e22', '#3498db', '#2ecc71', '#9b59b6', '#34495e']  # Professional colors
        )

        # Add legend with labels and values outside the pie chart
        ax.legend(
            [f"{cat}: {val}g" for cat, val in zip(categories, original_values)],
            title="Original Food Nutrition",
            loc="center left",
            bbox_to_anchor=(1, 0, 0.5, 1),
            fontsize=12,
            facecolor="white",
            edgecolor="black"
        )

        ax.set_title('Original Food Nutrition Distribution (Pie Chart)', fontsize=14)

        plt.tight_layout()
        plt.show()

        # Create a pie chart for the substituted food
        fig, ax = plt.subplots(figsize=(8, 8))  # Increased size for pie chart

        ax.pie(
            substitute_values,
            startangle=90,
            wedgeprops={'edgecolor': 'black', 'linewidth': 1},
            colors=['#f1c40f', '#e67e22', '#3498db', '#2ecc71', '#9b59b6', '#34495e']  # Professional colors
        )

        # Add legend with labels and values outside the pie chart
        ax.legend(
            [f"{cat}: {val}g" for cat, val in zip(categories, substitute_values)],
            title="Substitute Food Nutrition",
            loc="center left",
            bbox_to_anchor=(1, 0, 0.5, 1),
            fontsize=12,
            facecolor="white",
            edgecolor="black"
        )

        ax.set_title('Substitute Food Nutrition Distribution (Pie Chart)', fontsize=14)

        plt.tight_layout()
        plt.show()
    else:
        print("\nNo substitution was performed.")
        original_values = [original[cat] for cat in categories]

        # Create a pie chart for the original food
        fig, ax = plt.subplots(figsize=(8, 8))  # Increased size for pie chart

        ax.pie(
            original_values,
            startangle=90,
            wedgeprops={'edgecolor': 'black', 'linewidth': 1},
            colors=['#f1c40f', '#e67e22', '#3498db', '#2ecc71', '#9b59b6', '#34495e']  # Professional colors
        )

        ax.legend(
            [f"{cat}: {val}g" for cat, val in zip(categories, original_values)],
            title="Original Food Nutrition",
            loc="center left",
            bbox_to_anchor=(1, 0, 0.5, 1),
            fontsize=12,
            facecolor="white",
            edgecolor="black"
        )

        ax.set_title('Original Food Nutrition Distribution (Pie Chart)', fontsize=14)

        plt.tight_layout()
        plt.show()

# User interaction
print("Welcome to the Food Analysis System!")
food_name = input("Enter the food name: ")

ingredients, nutrition = fetch_nutrition(food_name)

if ingredients:
    print(f"\nIngredients: {ingredients}")
    print("\nNutritional Information (per serving):")
    for key, value in nutrition.items():
        print(f"{key.capitalize()}: {value}g")

    # Ask if user wants to see substitutes
    view_substitute = input("\nWould you like to see a substitute for this food? (yes/no): ").lower()

    if view_substitute == 'yes':
        substitutes, substitute_nutrition = substitute_ingredients(food_name)
        if substitutes:
            print(f"\nSubstitute Ingredients: {substitutes}")
            print("\nSubstitute Nutritional Information (per serving):")
            for key, value in substitute_nutrition.items():
                print(f"{key.capitalize()}: {value}g")
            # Display the comparison of nutrition values
            display_comparison(nutrition, substitute_nutrition)
        else:
            print("\nNo substitutes available.")
            display_comparison(nutrition)
    else:
        display_comparison(nutrition)
else:
    print("Food not found.")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import numpy as np

# Example: Replace these with your actual multi-class labels (10 classes: 0 to 9)
y_true = np.random.randint(0, 10, size=100)  # Randomly generated actual labels (0 to 9)
y_pred = np.random.randint(0, 10, size=100)  # Randomly generated predicted labels (0 to 9)

# Accuracy
accuracy = accuracy_score(y_true, y_pred)
print(f"Accuracy: {accuracy:.4f}")

# Precision (Macro-Averaged for Multi-Class)
precision = precision_score(y_true, y_pred, average='macro')  # 'macro' calculates metrics independently for each class
print(f"Precision (Macro-Averaged): {precision:.4f}")

# Recall (Macro-Averaged for Multi-Class)
recall = recall_score(y_true, y_pred, average='macro')  # 'macro' calculates metrics independently for each class
print(f"Recall (Macro-Averaged): {recall:.4f}")

# F1 Score (Macro-Averaged for Multi-Class)
f1 = f1_score(y_true, y_pred, average='macro')  # 'macro' calculates metrics independently for each class
print(f"F1 Score (Macro-Averaged): {f1:.4f}")

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
print("Confusion Matrix:")
print(cm)

# Display Confusion Matrix as a heatmap
cmd = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[f"Class {i}" for i in range(10)])
cmd.plot(cmap=plt.cm.Blues, xticks_rotation='vertical')
plt.title("Confusion Matrix (10 Classes)")
plt.show()
```
